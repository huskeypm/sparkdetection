```python
def myplot(img,fileName=None):
    plt.axis('equal')
    pcolormesh(img, cmap='gray')
    if fileName!=None:
        plt.gcf().savefig(fileName,dpi=300)
    #plt.axis('off')
    plt.tight_layout()
```

```python
def ReadImg(fileName):
    img = cv2.imread(fileName)
    if img is None:
        raise RuntimeError(fileName+" likely doesn't exist")

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return gray
```

```python
def GetSubsection(
  img,
  gimpCoords = np.array( [[532,482],[552,510]])):
  # [y,x]
  unitCell = img[gimpCoords[0,1]:gimpCoords[1,1],gimpCoords[0,0]:gimpCoords[1,0]]
  return unitCell
```

```python
def rotater(img, ang):
    rows,cols = img.shape
    M = cv2.getRotationMatrix2D((cols/2,rows/2),ang,1)
    dst = cv2.warpAffine(img,M,(cols,rows))
    return dst
```

```python
def GetPeriod(hstrip):    
    # thresh
    #print np.mean(hstrip)
    hstrip[np.where(hstrip - np.mean( hstrip ) < 0 )] = 0

    #plt.axis('equal')
    #pcolormesh(hstrip,cmap="gray")
    sumAxis = np.argmin( np.shape(hstrip)) # find smallest axis
    
    line = np.sum( hstrip, sumAxis)
    #plt.plot(line)#

    # identify peaks (leading edge) 
    thresh = np.array( line > np.mean(line),dtype=float)
    #plt.figure()
    #plt.plot(thresh)
    nPix = np.shape(thresh)[0]
    diff = thresh[1:]-thresh[0:(nPix-1)]
    leadingEdges = np.where(diff>0)[0] 
    totDist = np.array(leadingEdges[-1] - leadingEdges[0],dtype=float)
    nPeaks = np.shape( leadingEdges )[0]
    peakDist = np.int(totDist/(nPeaks-1))

    #plt.figure()
    #plt.plot(thresh)
    return peakDist
```

```python
def RocIt(image,superThresh,myThresh=600,nIter=200,doPlot=False):
    # solutions
    hits = np.zeros_like( image )

    # iterate
    superThreshCopy = np.copy(superThresh)
    nHits = 0
    while (nHits < nIter):
      isHit = hit.EvaluateNextHit(image,superThreshCopy,hits,
                                  innerMargin=7,outerMargin=13,threshold=myThresh,verbose=False)
      nHits += 1

      if isHit == False:
        # return on first false, since ppoints should be rank ordered
        break
    
    if doPlot:
        plt.figure()
        plt.axis('equal')
        plt.axis("off")
        pcolormesh(hits)
    
    return nHits 
```

```python
def padWithZeros(array, padwidth, iaxis, kwargs):
    array[:padwidth[0]] = 0
    array[-padwidth[1]:]= 0
    return array
```
