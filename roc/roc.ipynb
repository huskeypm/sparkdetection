```python
decoyImage = np.zeros_like(image)
decoyImage[:,220:265]=image[:,220:265]
plt.pcolormesh(decoyImage)
plt.colorbar()

```

```python

def GenThreshed(image,thresh=122):
    # record values above threshold 
    superThresh = np.copy(image)
    superThresh[image <= thresh]  = 0
    plt.figure()
    pcolormesh(superThresh)

    # sort by intensity 
    flat = np.ndarray.flatten(superThresh)
    sidx = np.argsort( flat )[::-1]
    plt.figure()
    plt.plot( flat[sidx])

    # plot up to first zero 
    firstZero = np.argwhere( flat[sidx]<1e-9)[0]
    plt.xlim([0,firstZero])

    return superThresh, sidx
```

```python
superThreshTruth,sidxTruth = GenThreshed(truthImage,thresh=thresh)
superThreshDecoy,sidxDecoy = GenThreshed(decoyImage,thresh=thresh)
```

#### Look for all points that exceed background of local environment (annulus)

```python
def RocIt(image,superThresh,myThresh=600,nIter=200,doPlot=False):
    # solutions
    hits = np.zeros_like( image )

    # iterate
    superThreshCopy = np.copy(superThresh)
    nHits = 0
    while (daIter < nIter):
      isHit = hit.EvaluateNextHit(image,superThreshCopy,hits,
                                  innerMargin=7,outerMargin=13,threshold=myThresh,verbose=False)
      nHits += 1

      if isHit == False:
        # return on first false, since ppoints should be rank ordered
        break
    
    if doPlot:
        plt.figure()
        plt.axis('equal')
        pcolormesh(hits)
    
    return nHits 
```
