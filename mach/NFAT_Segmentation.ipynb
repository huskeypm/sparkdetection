```python
import numpy as np
import cv2
```

```python
# load in data
root = "/home/AD/dfco222/Desktop/"
```

```python
#fileName = "NFAT_Despa_Fig6.png"
fileName = "NFAT_Despa_Fig6_2.png"
img = cv2.imread(root+fileName)
```

```python
imshow(img)
```

```python
if fileName == "NFAT_Despa_Fig6.png":
    img[0:70, 0:110] = 0
    imgDim = np.shape(img)
    img[(imgDim[0]-10):,:] = 0
    img[0:7,:] = 0
    img[:,0:4] = 0

if fileName == "NFAT_Despa_Fig6_2.png":
    img[0:40, 0:80] = 0



imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
imshow(gray)
plt.colorbar()
```

```python
if fileName == "NFAT_Despa_Fig6.png":
    retval, thresh = cv2.threshold(gray, 65, 255, cv2.THRESH_BINARY)

if fileName == "NFAT_Despa_Fig6_2.png":
    ## doing primary thresholding as to reduce amount of false positives introduced by gaussian adaptive thresholding
    threshParam = 46
    gray[gray <= threshParam] = 0
    imshow(gray)
    plt.colorbar()
    
    
    clahe = cv2.createCLAHE(clipLimit=1.0,tileGridSize=(3,3))
    cl1 = clahe.apply(gray)
    plt.figure()
    imshow(cl1)
    plt.colorbar()
    

    blockSize = 13
    thinningParameter = -4.
    thresh = cv2.adaptiveThreshold(cl1, 255, \
                                   cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
                                   #cv2.ADAPTIVE_THRESH_MEAN_C,\
                                    cv2.THRESH_BINARY, blockSize, 
                               thinningParameter)
    
    #retval, thresh = cv2.threshold(gray, 80, 255, cv2.THRESH_BINARY)
    
plt.figure()
imshow(thresh)
```

```python
write =  True#False#True
if write:
    cv2.imwrite(fileName[:-4]+"_thresholded.png", thresh)
```
