## segmentation of filtered image

using a gaussian kernel here to implement an adaptive threshold

```python
# h is between 0 and 1, must 'unnorm' the image to work with cv2
modifiedIMG = croppedIMG * 255 / np.max(croppedIMG)

util.myplot(modifiedIMG)
# setting noise = 0
modifiedIMG = np.floor(modifiedIMG)
figure()
util.myplot(modifiedIMG)

```

```python
## some values are negative so zero shifting
modifiedIMG = modifiedIMG - np.min(modifiedIMG)
# renorming
modifiedIMG = modifiedIMG * 255./np.max(modifiedIMG)

#modIMG = np.asarray(modifiedIMG)
modIMG = np.asarray(modifiedIMG, dtype=uint8)
#print modIMG
util.myplot(modIMG)



#kernelSize = 3

'''
It is important to understand what exactly is meant by thinning parameter.
This parameter is actually the value that is subtracted from the calculated weighted average of the x,y coords
of the modImg. By increasing the value, you are increasing the amount that is subtracted from this average.
This causes pixels on the outer edges of being considered within the threshold to be artificially percieved 
as lower than their actual threshold, and subsequently are given a value of 0 instead of 255.
'''
thinningParameter = 0

blockSize = 7

#hGray = cv2.GaussianBlur(h, (kernelSize,kernelSize), 0)
#plt.figure()
#imshow(hGray)

thresh = cv2.adaptiveThreshold(modIMG, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blockSize, 
                               thinningParameter)
#print type(hGray[1,1])
#print np.shape(hGray)
#thresh = cv2.adaptiveThreshold(hGray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blockSize, thinningParameter) 

plt.figure()
util.myplot(thresh)

```
