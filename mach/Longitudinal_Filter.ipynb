# Using methods for Ryan's ROC generation to pick out hits from ideal case
ideal case being the image the filter was constructed from (guo)

```python
def padWithZeros(array, padwidth, iaxis, kwargs):
    array[:padwidth[0]] = 0
    array[-padwidth[1]:]= 0
    return array
```

```python
import sys
sys.path.append("/home/AD/dfco222/scratchKafka/sparkdetection/roc/")
import hit
def RocIt(image,superThresh,myThresh=600,nIter=200,doPlot=False,innerAnnulusMargin=7, outerAnnulusMargin=13,verbose=False):
    # solutions
    hits = np.zeros_like( image )

    # iterate
    superThreshCopy = np.copy(superThresh)
    nHits = 0
    while (nHits < nIter):
      isHit = hit.EvaluateNextHit(image,superThreshCopy,hits,
                                  innerMargin=innerAnnulusMargin,outerMargin=outerAnnulusMargin,threshold=myThresh,verbose=verbose)
      nHits += 1
      #plt.figure()
      #util.myplot(superThreshCopy)
      if isHit == False:
        # return on first false, since ppoints should be rank ordered
        break
    
    if doPlot:
        plt.figure()
        plt.axis('equal')
        
        #plt.axis("off")
        pcolormesh(hits)
        plt.ylim([np.max(np.shape(hits)[0]),0])
    return nHits 
```
