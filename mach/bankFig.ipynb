# Basically just finds a 'unit cell' sized area around each detection 
# for the purpose of interpolating the data 
from scipy import signal
def doLabel(result,dx=10):
    img =result.stackedHits > 0
    kernel = np.ones((dx,dx),np.float32)/(dx*dx)
    
    filtered = signal.convolve(img, kernel, mode='same') / sum(kernel)

    plt.subplot(1,3,1)
    imshow(img)
    plt.subplot(1,3,2)
    imshow(filtered)
    plt.subplot(1,3,3)
    labeled = filtered > 0
    imshow(labeled)
    plt.tight_layout()
    
    return labeled
