# Matched Filter Tutorial Script

The purpose of this script is to demonstrate some key concepts of the matched filtering routine

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matchedFilter as mF
```

### Make Three Sizes of Images

```python
bigImg = np.zeros((300,300))
img = np.zeros((200,200))
smallImg = np.zeros((100,100))
```

```python
bigImg[150, 120:180] = 1
plt.figure()
imshow(bigImg)

img[100, 70:130] = 1
plt.figure()
imshow(img)

smallImg[50, 20:80] = 1
plt.figure()
imshow(smallImg)


imgDict = {}
imgDict['big'] = bigImg
imgDict['norm'] = img
imgDict['small'] = smallImg
```

### Make Three Sizes of Filters

```python
bigFilter = np.zeros((60,66))
bigFilter[29:31,3:-3] = 1
plt.figure()
imshow(bigFilter)

Filter = np.zeros((40, 66))
Filter[19:21,3:-3] = 1
plt.figure()
imshow(Filter)


smallFilter = np.zeros((20,66))
smallFilter[9:11,3:-3] = 1
plt.figure()
imshow(smallFilter)

filterDict = {}
filterDict['big'] = bigFilter
filterDict['norm'] = Filter
filterDict['small'] = smallFilter
```

```python
for size, img in imgDict.iteritems():
    for size2, myFilter in filterDict.iteritems():
        response = mF.matchedFilter(img, myFilter)
        plt.figure()
        imshow(response,cmap='gray')
        plt.title("img size "+size+" filter size "+size2)
        plt.colorbar()
        print "Image Size:", size, "\tFilter Size:", size2
        print "Minimum Response =", np.min(response), "\t", "Maximum Response =", np.max(response), "\t", "Average Response =", np.mean(response)
        print 120*"#", "\n"
```

### Testing filters that are longer and shorter

```python
longFilter = np.zeros((60,100))
longFilter[29:31,20:80] = 1
plt.figure()
imshow(longFilter)

medFilter = np.zeros((40, 80))
medFilter[19:21,10:70] = 1
plt.figure()
imshow(medFilter)


shortFilter = np.zeros((20,66))
shortFilter[9:11,3:-3] = 1
plt.figure()
imshow(shortFilter)

filterDictHoriz = {}
filterDictHoriz['long'] = longFilter
filterDictHoriz['med'] = medFilter
filterDictHoriz['short'] = shortFilter
```

```python
for size, img in imgDict.iteritems():
    for size2, myFilter in filterDictHoriz.iteritems():
        response = mF.matchedFilter(img, myFilter)
        plt.figure()
        imshow(response,cmap='gray')
        plt.title("img size "+size+" filter size "+size2)
        plt.colorbar()
        #print 50*"#"
        print "Image Size:", size, "\tFilter Size:", size2
        print "Minimum Response =", np.min(response), "\t", "Maximum Response =", np.max(response), "\t", "Average Response =", np.mean(response)
        print 120*"#", "\n"
```

### Conclusions From Above

 - It can be seen that the amount of blank space in the filter does not affect the response. 
 - It is also shown that the amount of blank space in the image does affect the response 
  - Due to the average response being defined as 0 (within computational limits).

```python
crowdedImg = imgDict['norm'].copy()
# bar is located at 100, so let's put some pixels to the upper right of it and see how each filter responds
# spacing bar so that it is out of the range of the small filter but still within range of larger filters
rowIndex = 100 - int(np.shape(filterDict['small'])[0]) - 5
print rowIndex
crowdedImg[rowIndex, 100:160] = 1
imshow(crowdedImg)
```

```python
for sizeFilter, Filter in filterDict.iteritems():
        response = mF.matchedFilter(crowdedImg, Filter)
        plt.figure()
        imshow(response,cmap='gray')
        plt.title("Filter size "+sizeFilter)
        plt.colorbar()
        #print 50*"#"
        #print "Image Size:", size, "\tFilter Size:", size2
        print "Filter Size:", sizeFilter
        print "Minimum Response =", np.min(response), "\t", "Maximum Response =", np.max(response), "\t", "Average Response =", np.mean(response)
        print "Sum of Response Along Row of Upper Bar =", np.sum(response[rowIndex])
        print "Maximum Response Along Row of Upper Bar =", np.max(response[rowIndex])
        print 120*"#", "\n"
```

### Conclusions (cont.)

 - The presence of the bar in the image did not affect the response of the matched filter, even when the bar to the upper right was within the range of the smallest filter
