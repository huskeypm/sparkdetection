# Matched Filter Tutorial Script

The purpose of this script is to demonstrate some key concepts of the matched filtering routine

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matchedFilter as mF
```

### Make Three Sizes of Images

```python
bigImg = np.zeros((300,300))
img = np.zeros((200,200))
smallImg = np.zeros((100,100))
```

```python
bigImg[150, 120:180] = 1
plt.figure()
imshow(bigImg)

img[100, 70:130] = 1
plt.figure()
imshow(img)

smallImg[50, 20:80] = 1
plt.figure()
imshow(smallImg)


imgDict = {}
imgDict['big'] = bigImg
imgDict['norm'] = img
imgDict['small'] = smallImg
```

### Make Three Sizes of Filters

```python
bigFilter = np.zeros((60,66))
bigFilter[29:31,3:-3] = 1
plt.figure()
imshow(bigFilter)

Filter = np.zeros((40, 66))
Filter[19:21,3:-3] = 1
plt.figure()
imshow(Filter)


smallFilter = np.zeros((20,66))
smallFilter[9:11,3:-3] = 1
plt.figure()
imshow(smallFilter)

filterDict = {}
filterDict['big'] = bigFilter
filterDict['norm'] = Filter
filterDict['small'] = smallFilter
```

```python
for size, img in imgDict.iteritems():
    for size2, myFilter in filterDict.iteritems():
        response = mF.matchedFilter(img, myFilter,demean=False)
        plt.figure()
        imshow(response,cmap='gray')
        plt.title("img size "+size+" filter size "+size2)
        plt.colorbar()
        print "Image Size:", size, "\tFilter Size:", size2
        print "Minimum Response =", np.min(response), "\t", "Maximum Response =", np.max(response), "\t", "Average Response =", np.mean(response)
        print 120*"#", "\n"
```

### Testing filters that are longer and shorter

```python
longFilter = np.zeros((60,100))
longFilter[29:31,20:80] = 1
plt.figure()
imshow(longFilter)

medFilter = np.zeros((40, 80))
medFilter[19:21,10:70] = 1
plt.figure()
imshow(medFilter)


shortFilter = np.zeros((20,66))
shortFilter[9:11,3:-3] = 1
plt.figure()
imshow(shortFilter)

filterDictHoriz = {}
filterDictHoriz['long'] = longFilter
filterDictHoriz['med'] = medFilter
filterDictHoriz['short'] = shortFilter
```

```python
for size, img in imgDict.iteritems():
    for size2, myFilter in filterDictHoriz.iteritems():
        response = mF.matchedFilter(img, myFilter,demean=False)
        plt.figure()
        imshow(response,cmap='gray')
        plt.title("img size "+size+" filter size "+size2)
        plt.colorbar()
        #print 50*"#"
        print "Image Size:", size, "\tFilter Size:", size2
        print "Minimum Response =", np.min(response), "\t", "Maximum Response =", np.max(response), "\t", "Average Response =", np.mean(response)
        print 120*"#", "\n"
```

### Conclusions From Above

 - It can be seen that the amount of blank space in the filter and image does not affect the response. 
 - When demeaning and parsevals are turned off, the filter response is the integrated area of the filter. (Bar is 60 pixels long in both cases, so the maximum filter response is 60)

```python
crowdedImg = imgDict['norm'].copy()
# bar is located at 100, so let's put some pixels to the upper right of it and see how each filter responds
# spacing bar so that it is out of the range of the small filter but still within range of larger filters
rowIndex = 100 - int(np.shape(filterDict['small'])[0]) - 5
print rowIndex
crowdedImg[rowIndex, 100:160] = 1
imshow(crowdedImg)
```

```python
for sizeFilter, Filter in filterDict.iteritems():
        response = mF.matchedFilter(crowdedImg, Filter,demean=False)
        plt.figure()
        imshow(response,cmap='gray')
        plt.title("Filter size "+sizeFilter)
        plt.colorbar()
        #print 50*"#"
        #print "Image Size:", size, "\tFilter Size:", size2
        print "Filter Size:", sizeFilter
        print "Minimum Response =", np.min(response), "\t", "Maximum Response =", np.max(response), "\t", "Average Response =", np.mean(response)
        print "Sum of Response Along Row of Upper Bar =", np.sum(response[rowIndex])
        print "Maximum Response Along Row of Upper Bar =", np.max(response[rowIndex])
        print 120*"#", "\n"
```

### Conclusions (cont.)

 - The presence of the bar in the image did not affect the response of the matched filter, even when the bar to the upper right was within the range of the smallest filter so matched filtering does not penalize the existance of blank space within the image, it just shows correlation

### Let's see how the filter responds to an image that now has background response

```python
backgroundImg = imgDict['norm'].copy()
backgroundImg += 10
plt.figure()
imshow(backgroundImg)
plt.colorbar()
for sizeFilter, Filter in filterDict.iteritems():
        response = mF.matchedFilter(backgroundImg, Filter,demean=False)
        plt.figure()
        imshow(response,cmap='gray')
        plt.title("Filter size "+sizeFilter)
        plt.colorbar()
        #print 50*"#"
        #print "Image Size:", size, "\tFilter Size:", size2
        print "Filter Size:", sizeFilter
        print "Minimum Response =", np.min(response), "\t", "Maximum Response =", np.max(response), "\t", "Average Response =", np.mean(response)
        print 120*"#", "\n"
```

### Conclusions (cont.)
- The response is still the sum of the area of the bar above the background response. So we're still good

```python
intenseImg = imgDict['norm'].copy() * 200.
plt.figure()
imshow(intenseImg)
plt.title("Raw Image")
plt.colorbar()
for sizeFilter, Filter in filterDict.iteritems():
        response = mF.matchedFilter(intenseImg, Filter,demean=False)
        plt.figure()
        imshow(response,cmap='gray')
        plt.title("Filter size "+sizeFilter)
        plt.colorbar()
        #print 50*"#"
        #print "Image Size:", size, "\tFilter Size:", size2
        print "Filter Size:", sizeFilter
        print "Minimum Response =", np.min(response), "\t", "Maximum Response =", np.max(response), "\t", "Average Response =", np.mean(response)
        print 120*"#", "\n"
```

### Conclusions (cont.)
- Just demonstrating that the intensity of the data (200/px in this case) multiplied by the intensity of filter (60 over the entire bar) since there is a perfect 1:1 correlation over the entire bar

```python
imgRows, imgCols = np.shape(imgDict['norm'])

lowNoise = np.random.rand(imgRows,imgCols) * 10.
lowNoise[np.where(imgDict['norm'] > 0)] = 100
plt.figure()
imshow(lowNoise)
plt.colorbar()

medNoise = np.random.rand(imgRows, imgCols) * 50.
medNoise[np.where(imgDict['norm'] > 0)] = 100
plt.figure()
imshow(medNoise)
plt.colorbar()

highNoise = np.random.rand(imgRows,imgCols) * 100.
highNoise[np.where(imgDict['norm'] > 0)] = 100
plt.figure()
imshow(highNoise)
plt.colorbar()

noiseDict = {}
noiseDict['low'] = lowNoise
noiseDict['medium'] = medNoise
noiseDict['high'] = highNoise
```

```python
for noise, img in noiseDict.iteritems():
    for sizeFilter, Filter in filterDict.iteritems():
        response = mF.matchedFilter(img, Filter,demean=False)
        plt.figure()
        imshow(response,cmap='gray')
        plt.title("Noise:"+noise+" Filter:"+sizeFilter)
        plt.colorbar()
        #print 50*"#"
        #print "Image Size:", size, "\tFilter Size:", size2
        print "Noise Content in Image:", noise, "\tFilter Size:", sizeFilter
        print "Minimum Response =", np.min(response), "\t", "Maximum Response =", np.max(response), "\t", "Average Response =", np.mean(response)
        print 120*"#", "\n"
```

### Conclusions (cont.)
- The filter responds to noise. (It should!)
