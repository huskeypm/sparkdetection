## Processing


#### Document Two Sarcomere Size of Images - EYEBALLED

```python
twoSarcSizeDict = {}
twoSarcSizeDict["Chen2015"] = 19.
twoSarcSizeDict["Guo2013Fig6A"] = 20.
twoSarcSizeDict["Guo2013Fig4C"] = 40.
twoSarcSizeDict["Guo2013Fig1C"] = 21.
```

#### Load in Normalization Images

```python
normalizationRoot = "./images/WT/"
#fileNames = []
normalImgs = {}
#imgHeights = {}
heightSarcLengthRatios = {}

import os
for fileName in os.listdir(normalizationRoot):
    # read in image
    img = util.ReadImg(normalizationRoot+fileName, renorm=True)
    imgHeight = np.shape(img)[0]
        
    #find filename minus the other garbage
    goodFileName = fileName.split("_")[0]
    
    # add to dictionary
    normalImgs[goodFileName] = img
    heightSarcLengthRatios[goodFileName] = imgHeight / twoSarcSizeDict[goodFileName]
    print imgHeight / twoSarcSizeDict[goodFileName]
    
    # add image dimensions to list
    #imgHeights[goodFileName] = imgHeight
    
    plt.figure()
    util.myplot(img)
    plt.title(goodFileName)
```

#### Ensure That Myocytes Are Represented in the Correct Aspect Ratio

```python
#print np.min(imgHeights)

#exDict = {'cool':5, 'ya':7, 'meh':1}
#min(exDict)

# find minimum height/twoSarcLength value
minRatio = min(heightSarcLengthRatios.itervalues())
print minRatio
#print minRatio

# cut down all images to that ratio
rightRatioImgs = {}
for name,img in normalImgs.iteritems():
    #if name != min(heightSarcLengthRatios):
    if True:
        # find middle row from image height
        imgHeight = np.shape(img)[0]
        midRow = round(float(imgHeight) / 2.)
                
        # assign new img
        newImg = img
        
        # grab all rows within range defined by minimum ratio already found
        rowRange = [int(midRow - round((minRatio*twoSarcSizeDict[name])/2.)),\
                    int(midRow + round((minRatio*twoSarcSizeDict[name])/2.))]
        print rowRange
        print np.shape(img)
        newImg = newImg[rowRange[0]:rowRange[1], :]
    else:
        newImg = img
    plt.figure()
    util.myplot(newImg)
    rightRatioImgs[name] = newImg
```

#### Resize the Images S.T. They All Have Same Two Sarcomere Distance

```python
#filterTwoSarcSize = 30.
filterTwoSarcSize = 24.
resizedImgs = {}
resizedHeights = {}
resizedLengths = {}
for key,img in rightRatioImgs.iteritems():
    scale = filterTwoSarcSize / twoSarcSizeDict[key] 
    resizedImgs[key] = cv2.resize(img, None, fx=scale, fy=scale, interpolation = cv2.INTER_CUBIC)
    resizedHeights[key] = np.shape(resizedImgs[key])[0]
    resizedLengths[key] = np.shape(resizedImgs[key])[1]
    plt.figure()
    util.myplot(resizedImgs[key])
    plt.title(key)
```

#### Double Checking To Make Sure Heights Are Roughly The Same

```python
stdDevHeights = np.std(resizedHeights.values())
minHeight = np.min(resizedHeights.values())
if stdDevHeights >= 5:
    raise RuntimeError, "Double check your two sarcomere size. You're getting a high standard deviation in the heights of the resized images."
```

#### Normalize Response

How about I make the prototypcial healthy myocyte besides one single longitudinal spot and one spot that has complete loss of TTs. This should then represent the full spectrum of responses I would get, right?

```python
jankyMyocyte = mashedMyocyte.copy()
jankDims = np.shape(jankyMyocyte)

# normalize
jankyMyocyte /= np.max(jankyMyocyte)

# simulate a complete loss of TTs in area that encompasses filter (upper left)
jankyMyocyte[:filterDim[0],:filterDim[1]] = 0

# simulate area that represents a 100% filter response (lower right)
jankyMyocyte[(jankDims[0]-filterDim[0]):,(jankDims[1]-filterDim[1]):] = truthFilter
jankyMyocyte /= np.max(jankyMyocyte)

imshow(jankyMyocyte)#,cmap='gray')
plt.title("Doctored Myocyte")
plt.colorbar()
```

```python
jankyH = mF.matchedFilter(jankyMyocyte,truthFilter,demean=False)

imshow(jankyH)
plt.title("Unaltered Filter Response")
plt.colorbar()

# just highlighting max value
jankyH[jankyH > (np.max(jankyH) * .99)] = 3.*np.max(jankyH)
plt.figure()
plt.title("Everything > 0.99 * max response, shaded")
imshow(jankyH)
plt.colorbar()
```

#### Well...

So my idea didn't work. The filter response of the filter image isn't even 0.9 * max response?? What now?

Could I just artificially insert a blank region simulating TT loss and a region representing the filter multiplied by some arbitrary value, then normalize from there?

#### Let's try it

```python
jankyMyocyte2 = mashedMyocyte.copy()
jankDims2 = np.shape(jankyMyocyte2)

# normalize
jankyMyocyte2 /= np.max(jankyMyocyte2)

# simulate a complete loss of TTs in area that encompasses filter (upper left)
jankyMyocyte2[:filterDim[0],:filterDim[1]] = 0

# I'm multiplying the filter here by an arbitrary value to ensure it is the max response in the filtered image
jankyMyocyte2[(jankDims2[0]-filterDim[0]):,(jankDims2[1]-filterDim[1]):] = truthFilter * 1.1 
jankyMyocyte2 /= np.max(jankyMyocyte2)

imshow(jankyMyocyte2)#,cmap='gray')
plt.title("Doctored Myocyte")
plt.colorbar()
```

```python
jankyH2 = mF.matchedFilter(jankyMyocyte2,truthFilter,demean=False)
plt.figure()
imshow(jankyH2)
plt.title("Raw Filter Response")
plt.colorbar()

normedJank = jankyH2 + abs(np.min(jankyH2))
normedJank /= np.max(normedJank)
plt.figure()
imshow(normedJank)
plt.title("Hand Wavy Results")
plt.colorbar()

# let's highlight the max response again
normedJank[normedJank == np.max(normedJank)] = 3
plt.figure()
imshow(normedJank,cmap='gray')
plt.title("Highlighted Max Response")
plt.colorbar()
```
