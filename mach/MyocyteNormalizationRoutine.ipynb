# Normalization Routine for Myocyte Matched Filter Response

Purpose of this script is to flesh out how exactly we can normalize myocyte filter response to a baseline response of the filter on the prototypical WT myoycte. This script will eventually be added to the util.py file located in the same directory

```python
import util
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [8,6]
%load_ext autoreload
%autoreload 2
import matchedFilter as mF
import mach
```

## Pre-processing

- Gather WT images that are clear and well imaged
- Process them s.t. the images is just the myocyte (no sarcolemma)
- Measure the two sarcomere distance of the myocyte
- Add the images to the normalization/WT repo


## Processing


#### Document Two Sarcomere Size of Images - EYEBALLED

```python
twoSarcSizeDict = {}
twoSarcSizeDict["Chen2015"] = 19.
twoSarcSizeDict["Guo2013Fig6A"] = 20.
twoSarcSizeDict["Guo2013Fig4C"] = 40.
twoSarcSizeDict["Guo2013Fig1C"] = 21.
```

#### Load in Normalization Images

```python
normalizationRoot = "./images/WT/"
#fileNames = []
normalImgs = {}
#imgHeights = {}
heightSarcLengthRatios = {}

import os
for fileName in os.listdir(normalizationRoot):
    # read in image
    img = util.ReadImg(normalizationRoot+fileName, renorm=True)
    imgHeight = np.shape(img)[0]
        
    #find filename minus the other garbage
    goodFileName = fileName.split("_")[0]
    
    # add to dictionary
    normalImgs[goodFileName] = img
    heightSarcLengthRatios[goodFileName] = imgHeight / twoSarcSizeDict[goodFileName]
    print imgHeight / twoSarcSizeDict[goodFileName]
    
    # add image dimensions to list
    #imgHeights[goodFileName] = imgHeight
    
    plt.figure()
    util.myplot(img)
    plt.title(goodFileName)
```

### Inverting Guo 2013Fig 4 C and Fig1C To Better Line Up Myocytes

```python
plt.figure()
imshow(normalImgs['Guo2013Fig4C'])
plt.title("Guo2013Fig4C")

normalImgs['Guo2013Fig4C'] = normalImgs['Guo2013Fig4C'][::-1,:]

plt.figure()
imshow(normalImgs['Guo2013Fig4C'])
plt.title("Guo2013Fig4C Inverted")

plt.figure()
imshow(normalImgs['Guo2013Fig1C'])
plt.title('Guo2013Fig1C')

normalImgs['Guo2013Fig1C'] = normalImgs['Guo2013Fig1C'][::-1,:]
plt.figure()
imshow(normalImgs['Guo2013Fig1C'])
plt.title('Guo2013Fig1C Inverted')
```

#### Ensure That Myocytes Are Represented in the Correct Aspect Ratio

```python
#print np.min(imgHeights)

#exDict = {'cool':5, 'ya':7, 'meh':1}
#min(exDict)

# find minimum height/twoSarcLength value
minRatio = min(heightSarcLengthRatios.itervalues())
print minRatio
#print minRatio

# cut down all images to that ratio
rightRatioImgs = {}
for name,img in normalImgs.iteritems():
    #if name != min(heightSarcLengthRatios):
    if True:
        # find middle row from image height
        imgHeight = np.shape(img)[0]
        midRow = round(float(imgHeight) / 2.)
                
        # assign new img
        newImg = img
        
        # grab all rows within range defined by minimum ratio already found
        rowRange = [int(midRow - round((minRatio*twoSarcSizeDict[name])/2.)),\
                    int(midRow + round((minRatio*twoSarcSizeDict[name])/2.))]
        print rowRange
        print np.shape(img)
        newImg = newImg[rowRange[0]:rowRange[1], :]
    else:
        newImg = img
    plt.figure()
    util.myplot(newImg)
    rightRatioImgs[name] = newImg
```

#### Resize the Images S.T. They All Have Same Two Sarcomere Distance

```python
#filterTwoSarcSize = 30.
filterTwoSarcSize = 24.
resizedImgs = {}
resizedHeights = {}
resizedLengths = {}
for key,img in rightRatioImgs.iteritems():
    scale = filterTwoSarcSize / twoSarcSizeDict[key] 
    resizedImgs[key] = cv2.resize(img, None, fx=scale, fy=scale, interpolation = cv2.INTER_CUBIC)
    resizedHeights[key] = np.shape(resizedImgs[key])[0]
    resizedLengths[key] = np.shape(resizedImgs[key])[1]
    plt.figure()
    util.myplot(resizedImgs[key])
    plt.title(key)
```

#### Double Checking To Make Sure Heights Are Roughly The Same

```python
stdDevHeights = np.std(resizedHeights.values())
minHeight = np.min(resizedHeights.values())
if stdDevHeights >= 5:
    raise RuntimeError, "Double check your two sarcomere size. You're getting a high standard deviation in the heights of the resized images."
```

#### Since Heights are Reasonably Close, Just Going to Chop Off Tops of Images to Make Them Same Heights

```python
for name, img in resizedImgs.iteritems():
    resizedImgs[name] = img[:minHeight, :]
    resizedHeights[name] = minHeight
```

#### Average Pixel Intensity Along Axial Direction

Thought being that if I collapse along axial direction and apply reasonable 'threshold' I can line up the peaks of the averages of each image s.t. the TTs line up and create a reasonable picture of a WT myocyte that accounts for some variation

```python
import operator
# sort dictionary into list of lengths of images in ascending order 
sortedLengths = sorted(resizedLengths.items(), key=operator.itemgetter(1))
print sortedLengths
```

```python
collapsedImgs = {}
for name,img in resizedImgs.iteritems(): 
    # average pixel intensity in axial direction
    avgPxIntensity = np.mean(img,axis=0)
    numCols = range(np.shape(img)[1])
    
    # applying 'threshold'
    avgPxIntensity[avgPxIntensity < np.mean(avgPxIntensity)] = 0
    
    print "numCols", numCols[-1]
    print "intensity length", np.shape(avgPxIntensity)
    plt.figure()
    #imshow(avgPxIntensity)
    plt.plot(numCols,avgPxIntensity)
    plt.title(name+" Averaged, Thresholded Intensity")
    plt.xlim([0,sortedLengths[-1][1]])
    collapsedImgs[name] = avgPxIntensity
```

Now we line up each image with the one that is closest to its length in ascending order, up to the top image. This should line up all of the images reasonably

```python
minLength = sortedLengths[0][1]
#print minLength
bestIdxs = []
for i,(imgName, length) in enumerate(sortedLengths[:-1]):
    nextImgName, nextLength = sortedLengths[i+1]
    
    # loop over img that is just longer than the current image
    squareErrorOld = 1e10 # arbitrarily large value
    
    for idx in range(0, (nextLength-minLength)): # want to use min length here so each image is fit to the final size
        try:
            err = resizedImgs[nextImgName][:,idx:(minLength+idx)] - resizedImgs[imgName]
        except:
            print "Images aren't the same dimension"
            print i
            break
        squareErrorNew = np.sum(np.power(err,2))
        if squareErrorNew < squareErrorOld:
            bestIdx = idx
            squareErrorOld = squareErrorOld
    bestIdxs.append(bestIdx)
    
    # changing image above so that all images will have same length
    resizedImgs[nextImgName] = resizedImgs[nextImgName][:,bestIdx:(bestIdx+minLength)]
print bestIdxs
```

```python
imgDims = []
for name,img in resizedImgs.iteritems():
    imgDim = np.shape(img)
    plt.figure()
    imshow(img)
    imgDims.append(imgDim)
for i in range(len(imgDims)):
    if imgDims[i] != imgDims[0]:
        raise RuntimeError, "Your images aren't the same dimension. Time to debug!"
```

#### Normalize Response

How about I make the prototypcial healthy myocyte besides one single longitudinal spot and one spot that has complete loss of TTs. This should then represent the full spectrum of responses I would get, right?

```python
jankyMyocyte = mashedMyocyte.copy()
jankDims = np.shape(jankyMyocyte)

# normalize
jankyMyocyte /= np.max(jankyMyocyte)

# simulate a complete loss of TTs in area that encompasses filter (upper left)
jankyMyocyte[:filterDim[0],:filterDim[1]] = 0

# simulate area that represents a 100% filter response (lower right)
jankyMyocyte[(jankDims[0]-filterDim[0]):,(jankDims[1]-filterDim[1]):] = truthFilter
jankyMyocyte /= np.max(jankyMyocyte)

imshow(jankyMyocyte)#,cmap='gray')
plt.title("Doctored Myocyte")
plt.colorbar()
```

```python
jankyH = mF.matchedFilter(jankyMyocyte,truthFilter,demean=False)

imshow(jankyH)
plt.title("Unaltered Filter Response")
plt.colorbar()

# just highlighting max value
jankyH[jankyH > (np.max(jankyH) * .99)] = 3.*np.max(jankyH)
plt.figure()
plt.title("Everything > 0.99 * max response, shaded")
imshow(jankyH)
plt.colorbar()
```

#### Let's try it

```python
jankyMyocyte2 = mashedMyocyte.copy()
jankDims2 = np.shape(jankyMyocyte2)

# normalize
jankyMyocyte2 /= np.max(jankyMyocyte2)

# simulate a complete loss of TTs in area that encompasses filter (upper left)
jankyMyocyte2[:filterDim[0],:filterDim[1]] = 0

# I'm multiplying the filter here by an arbitrary value to ensure it is the max response in the filtered image
jankyMyocyte2[(jankDims2[0]-filterDim[0]):,(jankDims2[1]-filterDim[1]):] = truthFilter * 1.1 
jankyMyocyte2 /= np.max(jankyMyocyte2)

imshow(jankyMyocyte2)#,cmap='gray')
plt.title("Doctored Myocyte")
plt.colorbar()
```

```python
jankyH2 = mF.matchedFilter(jankyMyocyte2,truthFilter,demean=False)
plt.figure()
imshow(jankyH2)
plt.title("Raw Filter Response")
plt.colorbar()

normedJank = jankyH2 + abs(np.min(jankyH2))
normedJank /= np.max(normedJank)
plt.figure()
imshow(normedJank)
plt.title("Hand Wavy Results")
plt.colorbar()

# let's highlight the max response again
normedJank[normedJank == np.max(normedJank)] = 3
plt.figure()
imshow(normedJank,cmap='gray')
plt.title("Highlighted Max Response")
plt.colorbar()
```
