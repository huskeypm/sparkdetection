## MACH

Mahalanobis, A., Kumar, B. V. K. V., Song, S., & Sims, S. R. F. (1994). Unconstrained correlation filters, 33(17), 3751–3759.

See eqn 18 (in my mendeley lib) 

$$
\begin{align}
h &= (D + S )^{-1} \langle x \rangle\\
D &\equiv \frac{1}{N} \sum_i Y_iY_i^*\quad\mbox{PSD of Y, fo which Y is member of False class}\\
S &\equiv \frac{1}{N} \sum_i (X_i-\langle x \rangle)^*(X_i-\langle x \rangle)
\end{align}
$$

$X_i$ is technically a $d\times d$ matrix, but we can instead use a rasterized image and therefore off-diagonal terms are zero. In this way, $X_i$ can be used as a vector

### Prepare MACH terms

Compute

$$
h = (S+C)^{-1}x
$$
since S is diagonal, its inverse is trivial. Further, we use 'C' to ensure that denominator is never zero. Note that 'C' could be considered as the PSD of the noise


```python
def MACHFilter(imgs, Cgamma = 1e-2):
    #Prepare matrix of vectorized of FFT'd images
    X = util.CalcX(imgs)#,debug=True) 

    # <x>
    meanX = np.mean(np.abs(X),axis=0)

    # S = 1/N sum (xi-x)^* (xi-x)
    xis_x = X - meanX
    #print "max",np.min(xis_x)
    daprod = np.conj(xis_x)*xis_x
    S = np.mean(daprod,axis=0)
    #print np.shape(S)
    #print "SD",np.max(S)

    #plt.figure()
    #util.myplot(np.reshape(xis_x[15,:],np.shape(img)))
    #util.myplot(np.reshape(S,np.shape(img)))
    
    
    # compute denom 
    C = np.ones(  np.prod(np.shape(S)))
    C*= Cgamma

    denom = C + S
    h = 1/denom * meanX
    
    
    # make 2D 
    H = np.reshape(h, np.shape(I))    
    return H

```
