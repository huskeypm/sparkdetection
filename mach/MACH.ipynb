## MACH

Based on (within some revision):
Mahalanobis, A., Kumar, B. V. K. V., Song, S., & Sims, S. R. F. (1994). Unconstrained correlation filters, 33(17), 3751–3759.

See eqn 18 (in my mendeley lib) 

$$
\begin{align}
h &= (D + S +C )^{-1} \langle x \rangle\\
D &\equiv \frac{1}{N} \sum_i Y_iY_i^*\quad\mbox{PSD of Y, for which Y is member of False class}\\
S &\equiv \frac{1}{N} \sum_i (X_i-\langle x \rangle)^*(X_i-\langle x \rangle)\\
C &= I\\
\end{align}
$$

Let $X_i$ represent the Fourier-transformed training image, for which N such images comprise $X$. 
Though $X_i$ is technically a $d\times d$ matrix, we can vectorize/rasterize the image (s.t. $X_i$ is a $d\times d$ length vector. 
In the math above, technically $X_i$ is still treated as a matrix, however we assume that the diagonal is represented by the rasterized image and therefore off-diagonal terms are zero. 
$\langle x \rangle$ is the mean image from $X$. 
The $D$, $S$, and $C$ terms can be thought of as the power spectral densities (PSDs) of decoy classes, dissimilarities among the training class, and the noise, respectively (see computation velow).
Given that $D$, $S$ and $C$ are diagonal, the inversion of $D+S+C$ is rendered trivial.

## Construction of large runner function

ImageDictionary - dictionary containing images with either 'true' or 'false' keys

TestImages - dict of images used to test if the filter is working. True images stored under 'true' key, false images under 'false'

```python
def main(imageDictionary, correlationThreshold=1.6e8):
    # Might wind up cutting out the reading of images from this func and instead write another func
    
    # Read in true images
    #TrueImages = []
    #[TrueImages.append(util.ReadImg(fileName,renorm=True,bound=ImageBounds)) for fileName in RawTrueImageLocations]
    #TrueImages = np.asarray(TrueImages,dtype=np.float)
    
    # Read in false images
    #FalseImages = []
    #[FalseImages.append(util.ReadImg(fileName,renorm=True,bound=[72,152])) for fileName in RawFalseImageLocations]
    #for fileName in RawFalseImageLocations:
    #    Img = util.ReadImg(fileName,renorm=True,bound=[72,152])
    #    FalseImages.append(Img)
    #FalseImages = np.asarray(FalseImages,dtype=np.float)
    trueImages = imageDictionary['true'][:]
    falseImages = imageDictionary['false'][:]
    testImages = imageDictionary['test'][:]
    '''
    Checking that all images are the same size
    May implement openCV to resize the images and circumvent this issue instead of using the image bounds method in util.py
    '''
    imgDim = np.shape(trueImages[0])
    assert( [((np.shape(image) == imgDim) for image in trueImages)]), "Make sure the images are the same size!" 
    
    # Create noisy image that is the same size as true and false images
    noiseLength, noiseHeight = imgDim
    noiseImg = np.reshape(np.random.randn(noiseLength*noiseHeight),[noiseLength,noiseHeight])
    noiseImg = util.renorm(noiseImg)
    
    # Make the filter
    truthFilter = mach.MACH2(trueImages)
    
    # Reading in test images
    #TrueTests = [util.ReadImg(TestImageDict['true'][i],renorm=True,bound=ImageBounds) 
    #             for i in range(np.shape(TestImageDict['true'])[0])]
    #FalseTests = [util.ReadImg(TestImageDict['false'][i],renorm=True,bound=ImageBounds) 
    #              for i in range(np.shape(TestImageDict['false'])[0])]

    
    # Testing filter/refining filter
    figure()
    for i,img in enumerate(testImages):
        img = util.renorm(img)
        corrTrue,idx,result = mach.fftcorr(img,truthFilter,correlationThreshold=correlationThreshold)
        assert(result), "This should be true. Fix your threshold, yo."
    
    for i,img in enumerate(falseImages):
        img = util.renorm(img)
        corrFalse,idx,result = mach.fftcorr(img,truthFilter,correlationThreshold=correlationThreshold)
        assert( result==False ), "This should be false. Fix your threshold, yo."
    
    noiseImg = util.renorm(noiseImg)
    corrFalse,idx,result = mach.fftcorr(noiseImg,truthFilter,correlationThreshold=correlationThreshold)
    assert( result==False ), "Your algorithm broke on the noisy image. Go grab some coffee, you'll need it."
    
    #Here would be a good spot to refine filter via simple while loop. 
    #(while result=False for True image, change threshold. Or something similar)
    print("An acceptable filter was generated, congrats!")
    return truthFilter
```
