```python
nPix =25
pore = np.ones([nPix,nPix])
idxs = np.arange(nPix)
mid = np.mean(idxs)

# Find stuff in middle and paint dark
xs = np.outer( idxs ,np.ones(nPix) )
ys = np.transpose(xs)
xd = xs - mid
yd = ys - mid
dist = xd**2 + yd**2
#print dist
cutoff = nPix/4.

idxs = np.where(dist < cutoff**2)
# stymied by the appropriate numpy way to do this. Dammit
pore[idxs]=0.

## Make it a ring for the heck of it
idxs = np.where(dist < (cutoff-2)**2)
# stymied by the appropriate numpy way to do this. Dammit
pore[idxs]=1.
```

```python
def myplot(img):
    plt.subplot(aspect='equal')
    plt.pcolormesh( img, cmap='gray' )
    
    
myplot(pore)    
```

### Make matrix of pore 

```python
def makeLattice(
    unitCell,  # img that's repeated 
    nX = 20,
    nY = 20,
    dislocate=False,
    bg = "ones"
    ):
    if bg=="zeros":
        lattice = np.zeros( [nX*nPix,nY*nPix] )
    else: 
        lattice = np.ones( [nX*nPix,nY*nPix] )
    for i in np.arange(nX):
        for j in np.arange(nY):
            idxi,idxj = i*nPix, j*nPix
            #print idxi,idxj
            
            
            if dislocate:
                #if i > int(nX/2.) or j > int(nY/2.):
                if i>j:
                    dim = np.array(np.shape(unitCell),dtype=float)
                    shift = [1*dim[0],0.5*dim[1]]
                        
                    #print shift    
                else:     
                    shift = [0,0]

                if i<nX-2 and j<nY-2:
                    lattice[(idxi+shift[0]):(idxi+nPix+shift[0]),
                        (idxj+shift[1]):(idxj+nPix+shift[1])] = unitCell
            else: 
                lattice[idxi:(idxi+nPix),idxj:(idxj+nPix)] = unitCell
            
    return lattice    
        
```

### Make lattices
- normal perfect
- dislocated

Images need to be inverted, s.t. signal is >0 

```python
lattice = makeLattice(pore)
# invert (e.g. signal is pore boundary)        
latticeInv = np.max(lattice)- lattice
poreInv  = np.max(pore)- pore

lattice = makeLattice(pore, dislocate=True)
# invert (e.g. signal is pore boundary)        
latticeDislocInv = np.max(lattice)- lattice

poreInv  = np.max(pore)- pore

plt.figure()
myplot( latticeInv )
plt.figure()
myplot( latticeDislocInv)

#plt.gcf().savefig("x.png",dpi=300)
```

```python
def addNoise(signal, amp=0.3):
    dims = np.shape(signal)
    noise = np.random.randn(np.prod(dims))
    noise = np.reshape(noise,dims,dims)
    amp=0.3
    measured = signal + amp * noise 
    return measured

measured = addNoise(latticeInv)
measuredDisloc = addNoise(latticeDislocInv)
plt.figure()
myplot( latticeDislocInv )
plt.figure()
myplot( measuredDisloc )
plt.figure()
myplot( poreInv)
```

```python
import matchedFilter as mF
h = mF.matchedFilter(measured,poreInv)
myplot(h)
```
