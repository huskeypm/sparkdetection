### More robust 

```python
def tester(testImages,truthFilter,color='k'):
    maxRats = []
    N, imgRows, imgCols = np.shape( testImages )
    for i in range(N):
        testImg = util.renorm( testImages[i], scale=1. )
        corrTrue,idx,result = mach.fftcorr(testImg,truthFilter,parsevals=True,verbose=False)
        plt.plot(corrTrue[idx[0],:],color)
        maxRat = np.max( corrTrue[idx[0],:] )/np.min(corrTrue[idx[0],:])
        maxRats.append(maxRat)

    return maxRats
    
trueRats = tester( imageDictionary['trueTrainers'] ,truthFilter,'k')        
print "False"
falseRats = tester( imageDictionary['falseTrainers'] ,truthFilter,'r')                
```

```python
cTrue,bins = np.histogram(trueRats)
#cFalse,dummy = np.histogram(falseRats,bins=bins)


plt.hist(trueRats,bins,alpha=0.5,label="True")
plt.hist(falseRats,bins,alpha=0.5,label="False")
plt.legend(loc=0)
```
