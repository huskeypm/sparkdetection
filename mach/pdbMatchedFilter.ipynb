```python


#########
import numpy as np
## load data
# have 3D array of points stored as 6x3 array (just reads columns )
coord = np.loadtxt("lig.pdb",usecols=[6,7,8])
print coord 

# retain x/y only
coord = coord[:,0:2]
damean =  np.mean(coord, axis=0)
coord -= damean
#print coord
```

```python
## assume each coordinate is a delta function (unit amplitude)
## so now we create a list where each entry is given a weight of 1.0
## to represent a single atom 
nC = np.shape(coord)[0] #  number of atoms
weights = np.ones(nC) # create array of ‘1.0s’ 

## However, we want to represent the data on a 2D regular mesh for the FFTs 
## later, so here we interpolate the ligand values (weights) onto 2D regular mesh
from scipy import mgrid
from scipy.interpolate import griddata
# make 2x2  grid with 10 points per interval
mins = np.min(coord,axis=0)
maxs = np.max(coord,axis=0)
gx,gy=mgrid[mins[0]:maxs[0]:100j,mins[1]:maxs[1]:100j]

## this is a bit of a hack, but I need to add zeros where the ligand is NOT found 
## I think of something better later....
empty = np.concatenate(\
  (coord+[0.1,0],coord-[0.1,0], coord+[0,0.1],coord-[0,0.1]),axis=0)
emptyweights = np.zeros(empty.shape[0]) # create zeros for each of these entries 
# concatenate
locs = np.concatenate((coord,empty),axis=0)
locweights = np.concatenate((weights,emptyweights),axis=0)

# interpolate 
d2D = griddata(locs,locweights, (gx,gy), method="nearest")  


## Display interpolated result 
plt.figure()
plt.gray()
plt.subplot(121,aspect="equal") 
plt.title("Original") 
plt.scatter(coord[:,0],coord[:,1])
plt.subplot(122,aspect="equal") 
plt.title("Interpolated")
plt.pcolormesh(d2D)
```

```python
#########

## First we'll smooth the ligand a bit so that we don't get strong Gibbs phenomena 
## we do this by creating a Gaussian filter that will be convolved against the ligand 
import scipy.signal as ss 
N = d2D.shape[0]
k = ss.gaussian(N,1) # 1D Gaussian (length N) with width of '1'
b = np.sqrt(np.outer(k,k)) # outer product with 1D array will give 2D array (use pcolormesh to view)
# need to shift by Nyquist so that Gaussian is centered at zero-frequency (so we're effectively doing a low-freq bandpass here) 
bs = fftp.fftshift(b) 

# smooth by convolving
d2Ds =fftp.ifft2(fftp.fft2(d2D)*fftp.fft2(bs))

# FFT representation
D2Ds = fftp.fft2( fftp.fftshift(d2Ds) ) 

## Display FFT result
plt.figure()
plt.gray()
plt.subplot(121,aspect="equal") 
plt.title("Smoothed ligand") 
plt.pcolormesh(d2Ds) 
plt.subplot(122,aspect="equal") 
plt.title("FFT representation\n (DC is at center)")
plt.pcolormesh(fftp.fftshift(D2Ds))


#####
```

### Repeat most everything, but do so using functions

```python
# Create a function based on the previous example to return the 
# ligand real-space representations
def GetLigand():
  coord = np.loadtxt("lig.pdb",usecols=[6,7,8])
  coord = coord[:,0:2]
  nC = np.shape(coord)[0] #  number of atoms
  damean =  np.mean(coord, axis=0)
  coord -= damean
    
  weights = np.ones(nC) # create array of 1s
  from scipy import mgrid
  from scipy.interpolate import griddata
  mins = np.min(coord,axis=0)
  maxs = np.max(coord,axis=0)
  gx,gy=mgrid[mins[0]:maxs[0]:100j,mins[1]:maxs[1]:100j]
  empty = np.concatenate((coord+[0.1,0],coord-[0.1,0], coord+[0,0.1],coord-[0,0.1]),axis=0)
  emptyweights = np.zeros(empty.shape[0]) # create zeros for each of these entries 
  locs = np.concatenate((coord,empty),axis=0)
  locweights = np.concatenate((weights,emptyweights),axis=0)
  d2D = griddata(locs,locweights, (gx,gy), method="nearest")  
  return (d2D)
```

```python

# create ligand 'image' 
d = GetLigand()

# create noisy test 'image' with ligand 
sImg = np.array([256,256]) # img size 
image = np.zeros(sImg)
sLig = np.shape(d)
image[0:sLig[0], 0:sLig[1]] = d  # embed ligand in image

# move the ligand around 
myshift = np.array([125,75])
image = np.roll(np.roll(image,myshift[0],axis=0),myshift[1],axis=1)
# make noise
noise = np.random.rand(np.product(sImg))
noise = np.reshape(noise,sImg)
noise = noise - np.mean(noise)
# add noise to images
image1 = image + 0.5*noise
image2 = image + 2*noise	# higher amplitude noise 
```

```python

## matched filter routine:
# Find ligand (filter) location in image
import scipy.fftpack as fftp
def matchedFilter(dimg,filter):
  # placeholder for 'noise' component (will refine later)
  fsC = np.ones(np.shape(dimg))
  
  ## prepare img
  # demean/shift img
  sdimg = fftp.fftshift(dimg - np.mean(dimg))
  # take FFT 
  fsdimg = fftp.fft2(sdimg)

  ## zero-pad filter
  si = np.shape(dimg)
  sf = np.shape(filter)
  # add zeros
  zeropad = np.zeros(si)
  zeropad[:sf[0],:sf[1]]=filter
  # shift original ligand by its Nyquist
  szeropad = np.roll(\
    np.roll(zeropad,-sf[0]/2+si[0]/2,axis=0),-sf[1]/2+si[1]/2,axis=1)
  f= szeropad

  ## signal
  # shift,demean filter
  sfilter = fftp.fftshift(f- np.mean(f))
  # take FFT
  fsfilter= fftp.fft2(sfilter)

  ## matched filter
  fsh = fsdimg * fsfilter / fsC
  sh = fftp.ifft2(fsh)
  h = fftp.ifftshift(sh)
  h = np.real(h)
  return h 
```

```python
## perform matched filter detections
cImg1 = matchedFilter(image1,d)
cImg2 = matchedFilter(image2,d) 
cNoise = matchedFilter(noise,d)

## plots
# top row has original images
# bottom row has correlation output 
plt.gray()
plt.subplot(2,3,1,aspect="equal")
plt.pcolormesh(image1)
plt.subplot(2,3,4,aspect="equal")
plt.pcolormesh(cImg1)
plt.subplot(2,3,2,aspect="equal")
plt.pcolormesh(image2)
plt.subplot(2,3,5,aspect="equal")
plt.pcolormesh(cImg2)
plt.subplot(2,3,3,aspect="equal")
plt.pcolormesh(noise)
plt.subplot(2,3,6,aspect="equal")
plt.pcolormesh(cNoise)
```

```python



########

# find the location and signal-to-noise ratio of the convolutions
def processDetection(corr,threshold):
  maxLoc = np.unravel_index(np.argmax( corr ),sImg) # find location of maximum 
  maxVal = np.max( corr )
  noisevariance = np.var(corr)  # note that we are 'polluting' our estimate of the noise variance with the 
                                                 #signal maximum, so a more careful approach would be to take an 
                                                 # annulus around the maximum
  SNR = maxVal / noisevariance
  if(SNR>threshold):
    print "Detection found at ", maxLoc, "with SNR ", SNR
  else:
    print "No detection" 

threshold = 2.5 # estimate from trial and error
processDetection(cImg1,threshold)
processDetection(cImg2,threshold)
processDetection(cNoise,threshold)
```
