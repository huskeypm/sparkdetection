```python
def GimmeBins(interiorContour, boxHeight, fullCheck=True):
    # chunks up contour into boxes. Edge boxes midpoints lie on the contour.
    # Interior contour generated from construct interior contour algorithm
    
    xCoords = []
    yCoords = []
    upperLeftPoint = []
    indexPoint = 0
    distOld = 100000
    #print np.shape(interiorContour)
    for i,point in enumerate(interiorContour):
        #saving for later
        x = point[0][0]
        y = point[0][1]
        xCoords.append(x)
        yCoords.append(y)
    
        distNew = sqrt(x**2 + y**2)
        if distNew < distOld:
            upperLeftPoint = (x,y)
            indexPoint = i
            distOld = distNew
    

    
    # stupid loop preconditioner
    leftX = np.min(xCoords)
    binnedXCoords = [leftX - boxHeight/2.]

    deltaX = boxHeight
    deltaY = deltaX
    binnedPoints = []
    i = 0
    while np.max(binnedXCoords) < (np.max(xCoords) - boxHeight/2.):

        indexes = np.where( np.asarray(map(lambda x: (x - (binnedXCoords[-1] + deltaX))**2, xCoords)) < 0.5 )[0]
    
        # take this out once debugged
        if np.shape(indexes) == (0,):
            break
    
        binnedX = xCoords[indexes[0]]
        binnedXCoords.append(binnedX)# index 0 is arbitrary since all indexes return same value of x
    
    
    
        # find where binnedX is
        indxs = np.where( np.asarray(map(lambda x: x - binnedX, xCoords)) == 0)[0]#xCoords == binnedX )

    
        # report all y values of binnedX
        maxY = np.max( [yCoords[index] for index in indxs]  )
        minY = np.min( [yCoords[index] for index in indxs]  )
    
        # append points to point array using pointPolygon checker method
        j = 0
        binnedYCoord = minY
        while binnedYCoord <= maxY:
        
            newBinnedPoint = (binnedX, binnedYCoord)
        
            # Gonna try testing all points to see if they are inside contour; meh, wasn't too much of a slow down
            fullCheck = True
            if fullCheck:
                if cv2.pointPolygonTest(interiorContour, newBinnedPoint, False) >= 0:
                    binnedPoints.append( newBinnedPoint )
            else:
                binnedPoints.append( newBinnedPoint )
            binnedYCoord += int(boxHeight)
            j += 1
        i += 1
    return binnedPoints
```

```python
def AnnulusResponse(binnedPoints, boxHeight, h, noiseInFilter):
    # Finds average SNR of filter response within inner box divided by SNR of outer box. Giving 'annulus' response.
    # it isn't really an annulus but whatever...
    binnedArrays = []
    filterResponse = np.zeros(imgDim)
    binAvgs = []
    binStdDevs = []
    SNRarray = np.zeros(imgDim)
    #filterArrays = []
    annulusResponses = []

    for point in binnedPoints:
        row = point[1]; col = point[0]
        #print row, col
        minRow = row - int(boxHeight/2.)
        maxRow = row + int(boxHeight/2.)
        rowRange = np.arange(minRow,maxRow)
    
        annulusMinRow = row - int(boxHeight/4.)
        annulusMaxRow = row + int(boxHeight/4.)
        annulusRowRange = np.arange(annulusMinRow,annulusMaxRow)
    
    
        minCol = col - int(boxHeight/2.)
        maxCol = col + int(boxHeight/2.)
        colRange = np.arange(minCol,maxCol)
    
        annulusMinCol = col - int(boxHeight/4.)
        annulusMaxCol = col + int(boxHeight/4.)
        annulusColRange = np.arange(annulusMinCol,annulusMaxCol)
    
        myPoints = []
        filterValues = []
        SNRvalues = []
        centerHit = 0
        centerPxlArea = np.shape(annulusRowRange)[0]*np.shape(annulusColRange)[0]
        outerHit = 0
        outerPxlArea = np.shape(rowRange)[0]*np.shape(colRange)[0] - centerPxlArea
        #if centerPxls+outerPxls != 100:
        #    raise RuntimeError
        for myRow in rowRange:
            for myCol in colRange:
                if myRow in annulusRowRange and myCol in annulusColRange:
                    # add in weighted filter hit
                    centerHit += h[myRow,myCol]/float(centerPxlArea)
                else:
                    outerHit += h[myRow,myCol]/float(outerPxlArea)
                myPoints.append( (myRow, myCol) )
                #print maskedImg[i,j]
                filterValues.append(h[myRow,myCol])
                filterResponse[myRow,myCol] = h[myRow,myCol]
                SNRarray[myRow,myCol] = h[myRow,myCol] / noiseInFilter
        filterValues = np.asarray(filterValues)
        #print filterValues
        #maxResponse = np.max(filterValues)
    
        annulusResponse = centerHit / outerHit
    
        #calculate avg and std dev of binned chunk
        binAvg = np.average(filterValues)
        binStdDev = np.std(filterValues)
    
        binnedArrays.append(myPoints) 
        #filterArrays.append(filterValues)
        binAvgs.append(binAvg)
        binStdDevs.append(binStdDev)
        #SNRarrays.append(SNRvalues)
        annulusResponses.append(annulusResponse)
    return binnedArrays, filterResponse, binAvgs, binStdDevs, SNRarray, annulusResponses
```

```python
def GimmePlots(binnedArrays, SNRarray, filterResponse, binAvgs, binStdDevs, annulusResponses=None):
    if annulusResponses != None:
        blankImg = np.zeros(imgDim)
        drawnImg = blankImg
        for i,myBin in enumerate(binnedArrays):
            response = annulusResponses[i]
            if response > 25:
                response = 25
            elif response < -1:
                response = -1
            for point in myBin:
                drawnImg[point[0],point[1]] = response
        plt.figure()
        imshow(drawnImg)
        plt.title("Annulus SNR")
        plt.colorbar()

    plt.figure()
    plt.imshow(filterResponse)
    plt.title("Filter Response")
    plt.colorbar()
    
    plt.figure()
    imshow(SNRarray)
    plt.title("SNR")
    plt.colorbar()
    
    blankImg = np.zeros(imgDim)
    drawnImg = blankImg
    for i,myBin in enumerate(binnedArrays):
        for point in myBin:
            drawnImg[point[0],point[1]] = binAvgs[i]
    plt.figure()
    imshow(drawnImg)
    plt.title("Average Filter Response of Bins")
    plt.colorbar()
    
    blankImg = np.zeros(imgDim)
    drawnImg = blankImg
    maxStdDev = np.max(binStdDevs)
    for i,myBin in enumerate(binnedArrays):
        for point in myBin:
            #if binStdDevs[i] < 3./4. * maxStdDev:
            #    drawnImg[point[0],point[1]] = binStdDevs[i]
            drawnImg[point[0],point[1]] = binStdDevs[i]
    plt.figure()
    imshow(drawnImg)
    plt.title("Bin Standard Deviations")
    plt.colorbar()
```
