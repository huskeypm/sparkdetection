```python
def SimpleBinningAndResponse(binnedPoints,boxHeight,RawFilterResponse):
    '''
    This function, given a list of binned points of form [ [column0, row0], [column1, row1], ... ]
        that make up the interior contour generated from the interior contour constructor function,
        a raw filter response from matched filtering (or any signal, rather), and a boxHeight;
        returns information based on the binning of the filter response.
    ''' 
    
    binnedArrays = []
    filterResponse = np.zeros(imgDim)
    binAvgs = []
    binStdDevs = []
    SNRarray = np.zeros(imgDim)
    SNRbinAvgs = []
    
    for point in binnedPoints:
        row = point[1]; col = point[0]
        #print row, col
        minRow = row - int(boxHeight/2.)
        maxRow = row + int(boxHeight/2.)
        rowRange = np.arange(minRow,maxRow)
        #print rowRange
    
        minCol = col - int(boxHeight/2.)
        maxCol = col + int(boxHeight/2.)
        colRange = np.arange(minCol,maxCol)
    
        myPoints = []
        filterValues = []
        SNRvalues = []
        for myRow in rowRange:
            for myCol in colRange:
                myPoints.append( (myRow, myCol) )
                #print maskedImg[i,j]
                filterValues.append(RawFilterResponse[myRow,myCol])
                filterResponse[myRow,myCol] = RawFilterResponse[myRow,myCol]
                SNRarray[myRow,myCol] = RawFilterResponse[myRow,myCol] / noiseInFilter
        filterValues = np.asarray(filterValues)
        #print filterValues
        maxResponse = np.max(filterValues)
    
    
        #calculate avg and std dev of binned chunk
        binAvg = np.average(filterValues)
        SNRbinAvg = binAvg / noiseInFilter
        binStdDev = np.std(filterValues)
    
        binnedArrays.append(myPoints) 
        #filterArrays.append(filterValues)
        binAvgs.append(binAvg)
        binStdDevs.append(binStdDev)
        #SNRarrays.append(SNRvalues)
        SNRbinAvgs.append(SNRbinAvg)
    return binnedArrays, filterResponse, binAvgs, binStdDevs, SNRarray, SNRbinAvgs
```

```python
def GimmeBins(interiorContour, boxHeight, fullCheck=True):

    xCoords = []
    yCoords = []
    upperLeftPoint = []
    indexPoint = 0
    distOld = 100000
    #print np.shape(interiorContour)
    for i,point in enumerate(interiorContour):
        #saving for later
        x = point[0][0]
        y = point[0][1]
        xCoords.append(x)
        yCoords.append(y)
    
        distNew = sqrt(x**2 + y**2)
        if distNew < distOld:
            upperLeftPoint = (x,y)
            indexPoint = i
            distOld = distNew
    

    
    # stupid loop preconditioner
    leftX = np.min(xCoords)
    binnedXCoords = [leftX - boxHeight/2.]

    deltaX = boxHeight
    deltaY = deltaX
    binnedPoints = []
    i = 0
    while np.max(binnedXCoords) < (np.max(xCoords) - boxHeight/2.):

        indexes = np.where( np.asarray(map(lambda x: (x - (binnedXCoords[-1] + deltaX))**2, xCoords)) < 0.5 )[0]
    
        # take this out once debugged
        if np.shape(indexes) == (0,):
            break
    
        binnedX = xCoords[indexes[0]]
        binnedXCoords.append(binnedX)# index 0 is arbitrary since all indexes return same value of x
    
    
    
        # find where binnedX is
        indxs = np.where( np.asarray(map(lambda x: x - binnedX, xCoords)) == 0)[0]#xCoords == binnedX )

    
        # report all y values of binnedX
        maxY = np.max( [yCoords[index] for index in indxs]  )
        minY = np.min( [yCoords[index] for index in indxs]  )
    
        # append points to point array using pointPolygon checker method
        j = 0
        binnedYCoord = minY
        while binnedYCoord <= maxY:
        
            newBinnedPoint = (binnedX, binnedYCoord)
        
            # Gonna try testing all points to see if they are inside contour; meh, wasn't too much of a slow down
            fullCheck = True
            if fullCheck:
                if cv2.pointPolygonTest(interiorContour, newBinnedPoint, False) >= 0:
                    binnedPoints.append( newBinnedPoint )
            else:
                binnedPoints.append( newBinnedPoint )
            binnedYCoord += int(boxHeight)
            j += 1
        i += 1
    return binnedPoints
```

```python
def AnnulusResponse(binnedPoints, boxHeight, h, noiseInFilter):
    '''
    This function, given a list of binned points of form [ [column0, row0], [column1, row1], ... ]
        that make up the interior contour generated from the interior contour constructor function,
        a raw filter response from matched filtering (or any signal, rather), and a boxHeight;
        returns information based on the binning of the filter response and the 'annulus' response.
        Annulus here really meaning concentric boxes. The average response within the interior box
        divided by the average response from the (exterior box area minus the interior box area).
    ''' 
    binnedArrays = []
    filterResponse = np.zeros(imgDim)
    binAvgs = []
    binStdDevs = []
    SNRarray = np.zeros(imgDim)
    #filterArrays = []
    annulusResponses = []
    SNRbinAvgs = []

    for point in binnedPoints:
        row = point[1]; col = point[0]
        #print row, col
        minRow = row - int(boxHeight/2.)
        maxRow = row + int(boxHeight/2.)
        rowRange = np.arange(minRow,maxRow)
    
        annulusMinRow = row - int(boxHeight/4.)
        annulusMaxRow = row + int(boxHeight/4.)
        annulusRowRange = np.arange(annulusMinRow,annulusMaxRow)
    
    
        minCol = col - int(boxHeight/2.)
        maxCol = col + int(boxHeight/2.)
        colRange = np.arange(minCol,maxCol)
    
        annulusMinCol = col - int(boxHeight/4.)
        annulusMaxCol = col + int(boxHeight/4.)
        annulusColRange = np.arange(annulusMinCol,annulusMaxCol)
    
        myPoints = []
        filterValues = []
        SNRvalues = []
        centerHit = 0
        centerPxlArea = np.shape(annulusRowRange)[0]*np.shape(annulusColRange)[0]
        outerHit = 0
        outerPxlArea = np.shape(rowRange)[0]*np.shape(colRange)[0] - centerPxlArea
        #if centerPxls+outerPxls != 100:
        #    raise RuntimeError
        for myRow in rowRange:
            for myCol in colRange:
                if myRow in annulusRowRange and myCol in annulusColRange:
                    # add in weighted filter hit
                    centerHit += h[myRow,myCol]/float(centerPxlArea)
                else:
                    outerHit += h[myRow,myCol]/float(outerPxlArea)
                myPoints.append( (myRow, myCol) )
                #print maskedImg[i,j]
                filterValues.append(h[myRow,myCol])
                filterResponse[myRow,myCol] = h[myRow,myCol]
                SNRarray[myRow,myCol] = h[myRow,myCol] / noiseInFilter
        filterValues = np.asarray(filterValues)
        #print filterValues
        #maxResponse = np.max(filterValues)
    
        annulusResponse = centerHit / outerHit
    
        #calculate avg and std dev of binned chunk
        binAvg = np.average(filterValues)
        SNRbinAvg = binAvg / noiseInFilter
        binStdDev = np.std(filterValues)
    
        binnedArrays.append(myPoints) 
        #filterArrays.append(filterValues)
        binAvgs.append(binAvg)
        binStdDevs.append(binStdDev)
        #SNRarrays.append(SNRvalues)
        annulusResponses.append(annulusResponse)
        SNRbinAvgs.append(SNRbinAvg)
    return binnedArrays, filterResponse, binAvgs, binStdDevs, SNRarray, annulusResponses, SNRbinAvgs
```

```python
def GimmePlots(binnedArrays, SNRarray, filterResponse, binAvgs, binStdDevs, SNRbinAvgs, annulusResponses=None):
    if annulusResponses != None:
        blankImg = np.zeros(imgDim)
        drawnImg = blankImg
        for i,myBin in enumerate(binnedArrays):
            response = annulusResponses[i]
            if response > 5:
                response = 5
            elif response < -1:
                response = -1
            for point in myBin:
                drawnImg[point[0],point[1]] = response
        plt.figure()
        imshow(drawnImg)
        plt.title("Annulus SNR")
        plt.colorbar()

    plt.figure()
    plt.imshow(filterResponse)
    plt.title("Filter Response")
    plt.colorbar()
    
    plt.figure()
    imshow(SNRarray)
    plt.title("SNR")
    plt.colorbar()
    
    blankImg = np.zeros(imgDim)
    drawnImg = blankImg
    for i,myBin in enumerate(binnedArrays):
        for point in myBin:
            drawnImg[point[0],point[1]] = binAvgs[i]
    plt.figure()
    imshow(drawnImg)
    plt.title("Average Filter Response of Bins")
    plt.colorbar()
    
    blankImg = np.zeros(imgDim)
    drawnImg = blankImg
    for i,myBin in enumerate(binnedArrays):
        for point in myBin:
            drawnImg[point[0],point[1]] = SNRbinAvgs[i]
    plt.figure()
    imshow(drawnImg)
    plt.title("Average SNR of Bins")
    plt.colorbar()
    
    blankImg = np.zeros(imgDim)
    drawnImg = blankImg
    maxStdDev = np.max(binStdDevs)
    for i,myBin in enumerate(binnedArrays):
        for point in myBin:
            #if binStdDevs[i] < 3./4. * maxStdDev:
            #    drawnImg[point[0],point[1]] = binStdDevs[i]
            drawnImg[point[0],point[1]] = binStdDevs[i]
    plt.figure()
    imshow(drawnImg)
    plt.title("Bin Standard Deviations")
    plt.colorbar()
```

```python
def GimmeAvgAndStdDev(rawFilterResponse, maskImg):
    filterResponses = []
    for i in range(np.shape(maskImg)[0]):
        for j in range(np.shape(maskImg)[1]):
            if maskImg[i,j] != 0:
                filterResponses.append(rawFilterResponse[i,j])
                #print h[i,j]
    filterResponses = np.asarray(filterResponses)
    
    stdDev = np.std(filterResponses)
    avg = np.mean(filterResponses)
    return avg, stdDev
```

```python
def CollapsedFilterResponse(rawFilterResponse,maskedImg):
    columnSums = []
    for col in range(np.shape(rawFilterResponse)[1]):
        columnSum = 0
        for row in range(np.shape(rawFilterResponse)[0]):
            if maskedImg[row,col] != 0:
                columnSum += rawFilterResponse[row,col]
        #columnSum = np.sum(rawFilterResponse[:,col])
        columnSums.append(columnSum)
    
    #for col in range(np.shape(maskedImg)[1]):
    #    for row in range(np.shape(maskedImg)[0]):
    #        if 
    
    columnFilterSize = np.sum(maskedImg,axis=0) / np.max(maskedImg)
    #print len(columnFilterSize), len(columnSums)
    
    columnAvgResponse = [columnSums[i] / columnFilterSize[i] for i in range(len(columnSums))]
    return columnSums, columnAvgResponse
```
